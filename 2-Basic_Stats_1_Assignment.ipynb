```python
#All the numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']

for column_name in numerical_columns:
    print('Mean calculation for', column_name)
    column_data = data[column_name]
    print('Sum of the values in the data is: ', sum(column_data))
    print('Count of the values in the data is: ', len(column_data))
    average_data = sum(column_data) / len(column_data)
    print('Mean of the values in the data is: ', round(average_data,2))
    print('')  # For better readability
```

```python
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']

for col_name in numerical_columns:
    print('Median calculation for: ', col_name)
    col_data = list(data[col_name].sort_values())
    print('The count of the data is: ', len(col_data))
    print('After sorting, the data is: ', col_data)

    indexvalue = len(col_data) // 2

    if len(col_data) % 2 == 0:  # Even number of elements
        print('The indexvalue is: ', indexvalue)
        median = (col_data[indexvalue - 1] + col_data[indexvalue]) / 2
    else:  # Odd number of elements
        print('The indexvalue is: ', indexvalue)
        median = col_data[indexvalue]

    print(f'Median for {col_name} is: ', median)
    print('')  # For better readability
```

**MODE**

```python
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

for col_name in numerical_columns:
    print(f'Mode calculation for {col_name}:')

    # Initialize lists
    unique_values = []
    frequency_list = []

    # Extract column data as a list
    column_data = list(data[col_name])

    # Find unique values
    for value in column_data:
        if value not in unique_values:
            unique_values.append(value)

    print(f'The unique values in the column {col_name}:', unique_values)

    # Count the frequency of each unique value
    for unique_value in unique_values:
        count = 0
        for value in column_data:
            if unique_value == value:
                count += 1
        frequency_list.append(count)

    print(f'Frequency of each value in the column {col_name}:', frequency_list)

    # Determine the highest frequency
    highest_frequency = max(frequency_list)
    print(f'Highest Frequency in {col_name}:', highest_frequency)

    # Print the mode(s)
    print(f'The mode(s) for {col_name} is/are:')

    for k in range(len(frequency_list)):
        if frequency_list[k] == highest_frequency:
            print(unique_values[k])

    print('')  # For better readability between columns
```

**STANDARD DEVIATION**

```python
import pandas as pd
import numpy as np

numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Initialize lists to store the results
columns = []
means = []
variances = []
std_devs = []

# Loop through each numerical column and calculate variance and standard deviation
for col_name in numerical_columns:

    # Step 1: Calculate the mean
    mean_value = data[col_name].mean()

    # Step 2: Calculate variance and standard deviation
    variance = data[col_name].var()
    std_dev = data[col_name].std()

    # Store the results in the lists
    columns.append(col_name)
    means.append(mean_value)
    variances.append(variance)
    std_devs.append(std_dev)

# Create a DataFrame to display the results in a table format
results_df = pd.DataFrame({
    'Column': columns,
    'Mean': means,
    'Variance': variances,
    'Standard Deviation': std_devs
})

# Display the table
print(results_df)
```

**MEAN :**

2.   A measure of central tendency.

3.   It helps understand where the center of the data lies but can be
     influenced by extreme values (outliers).

**MEDIAN :**

1.   The median is the middle number when all the numbers are arranged    in   order.

2.   Another measure of central tendency, less affected by outliers and skewed data.

3.   If you line up all the numbers from smallest to largest, the median is the number right in the middle. If there are an even number of values, it’s the average of the two middle numbers.


**MODE :**


1.   The mode is the value that occurs most frequently in the dataset. A dataset may have one mode, more than one mode, or no mode if no value repeats.

2.   The mode indicates the most common value(s) in the dataset. It is useful for understanding the most frequently occurring data point, especially in categorical or discrete data.

**STANDARD DEVIATION :**

1.  The standard deviation measures the dispersion or spread of the data values from the mean.

**Boxplots:**

```python
import pandas as pd
import matplotlib.pyplot as plt

numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Create boxplots for each numerical column
for col_name in numerical_columns:
    plt.figure(figsize=(8, 6))
    data.boxplot(column=col_name, vert=True, patch_artist= True, notch=True)
    plt.title(f'Boxplot for {col_name}')
    plt.xlabel(col_name)
    plt.grid(True)

    # Calculate quartiles and IQR
    Q1 = data[col_name].quantile(0.25)
    Q3 = data[col_name].quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col_name] < lower_bound) | (data[col_name] > upper_bound)][col_name]

    # Print statistics
    print(f'\nStatistics for {col_name}:')
    print(f'  Q1 (25th percentile): {Q1}')
    print(f'  Q3 (75th percentile): {Q3}')
    print(f'  IQR: {IQR}')
    print(f'  Lower Bound for Outliers: {lower_bound}')
    print(f'  Upper Bound for Outliers: {upper_bound}')
    print(f'  Number of Outliers: {len(outliers)}')
    if not outliers.empty:
        print(f'  Outliers: {outliers.tolist()}')

    plt.show()
```



1.   Outliers can indicate anomalies, errors in data, or unique cases that
     may need further investigation.

2.   A larger IQR suggests a wider spread in the central data values,
     whereas a smaller IQR indicates less variability.

3.   If you see multiple clusters of outliers or separated boxes, it might
     suggest multiple underlying distributions within the data.

#Standardization of Numerical Variables

**Objective:** To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

**Steps:**

Explain the concept of standardization (z-score normalization).

Standardize the numerical columns using the formula: z=x-mu/sigma



​Show before and after comparisons of the data distributions.

**Z-score normalization** refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

**New value** = (x – μ) / σ

**where:**

x: Original value
μ: Mean of data
σ: Standard deviation of data

```python
import pandas as pd
import numpy as np

numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Iterate over each column in the numerical_columns list
for column in numerical_columns:
    # Standardize each column's data individually
    data[column] = (data[column] - data[column].mean()) / data[column].std()

print("Standardized Data (Numerical Columns Only):")
print(data)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Plot original distributions
plt.figure(figsize=(12, 8))

# First row of plots
plt.subplot(2, 3, 1)
sns.histplot(data['Volume'], kde=True)
plt.title("Original Volume Distribution")

plt.subplot(2, 3, 2)
sns.histplot(data['Avg Price'], kde=True)
plt.title("Original Avg Price Distribution")

plt.subplot(2, 3, 3)
sns.histplot(data['Total Sales Value'], kde=True)
plt.title("Original Total Sales Value Distribution")

# Second row of plots
plt.subplot(2, 3, 4)
sns.histplot(data['Discount Rate (%)'], kde=True)
plt.title("Original Discount Rate (%) Distribution")

plt.subplot(2, 3, 5)
sns.histplot(data['Discount Amount'], kde=True)
plt.title("Original Discount Amount Distribution")

plt.subplot(2, 3, 6)
sns.histplot(data['Net Sales Value'], kde=True)
plt.title("Original Net Sales Value Distribution")


plt.show()

# Standardize the columns
for column in numerical_columns:
    data[column] = (data[column] - data[column].mean()) / data[column].std()

# Plot standardized distributions
plt.figure(figsize=(12, 8))

# First row of plots
plt.subplot(2, 3, 1)
sns.histplot(data['Volume'], kde=True)
plt.title("Standardized Volume Distribution")

plt.subplot(2, 3, 2)
sns.histplot(data['Avg Price'], kde=True)
plt.title("Standardized Avg Price Distribution")

plt.subplot(2, 3, 3)
sns.histplot(data['Total Sales Value'], kde=True)
plt.title("Standardized Total Sales Value Distribution")

# Second row of plots
plt.subplot(2, 3, 4)
sns.histplot(data['Discount Rate (%)'], kde=True)
plt.title("Standardized Discount Rate (%) Distribution")

plt.subplot(2, 3, 5)
sns.histplot(data['Discount Amount'], kde=True)
plt.title("Standardized Discount Amount Distribution")

plt.subplot(2, 3, 6)
sns.histplot(data['Net Sales Value'], kde=True)
plt.title("Standardized Net Sales Value Distribution")

plt.show()
```

When the histograms for both the original and standardized data look similar, it means that the shape of the distribution has not changed during standardization. This is expected because standardization (using Z-scores) does not alter the relative distribution or shape of the data—it only rescales it.

This just centers the data around zero, making it easier to compare different datasets or features.

#Conversion of Categorical Data into Dummy Variables

**Objective:** To transform categorical variables into a format that can be provided to ML algorithms.

**Steps:**

Discuss the need for converting categorical data into dummy variables (one-hot encoding).

Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

Display a portion of the transformed dataset.

```python
categorical_columns = data.select_dtypes(include=['object','category']).columns
categorical_columns
```

```python
import pandas as pd

# Identify categorical columns
categorical_column = ['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# Apply one-hot encoding to categorical columns
data_encoded = pd.get_dummies(data, columns=categorical_column, dtype='int')

# Convert the columns to integer type (usually done automatically, but for clarity)
for column in data_encoded.columns:
    if data_encoded[column].dtype == 'bool':
        data_encoded[column] = data_encoded[column].astype('int')

# Display the transformed dataset
print("Transformed Dataset with Dummy Variables:")
print(data_encoded)
```
